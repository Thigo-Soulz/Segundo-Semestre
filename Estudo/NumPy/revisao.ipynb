### 1)Considere que uma universidade registrou as notas de 100 alunos em 6 disciplinas, organizadas em uma matriz 2D, onde cada linha representa um aluno e cada coluna representa uma disciplina. Os valores das notas variam de 0 a 10.
### a) Utilizando a biblioteca NumPy, crie um array chamado notas com dimensão (100, 6), contendo valores aleatórios entre 0 e 10. Em seguida, exiba o shape, o número de dimensões (ndim) e o total de elementos (size) do array.

```python
import numpy as np 
import random as rd 

a = np.random.randint(0 ,11 , [100, 6])
print(a)
print(a.shape)
print(a.ndim)
print(a.size)
```

```python
b = a.reshape([50,12])
print(b)
b = a.reshape([100,6])
print(b)
```
