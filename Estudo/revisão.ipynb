```python
print(notas[1])

notas = notas + 0.5

print(notas[1])
```

```python
media = round(notas.mean())

print(media)

print(notas[1])

diff = notas - media

print(diff[1])
```

```python
print(notas.mean(axis=0)) #percorre colunas
print(notas.mean(axis=1)) #percorre linhas 
```

```python
print(notas.max(axis=1))
print(notas.std()) #desvio padrão
```

```python
maior = notas.max()

indices = np.argwhere(notas == maior)

print(indices)
```

```python
print(np.where(notas >= 7,True,False))
```

```python
medias = notas.mean(axis=1)

medias = medias[medias>=7]

print(medias.size)
```

```python
grupo_a = np.array([0,1,2,3,4])
grupo_b = np.array([3,4,5,6,7])

#inteseção
print(np.intersect1d(grupo_a,grupo_b))

#união
print(np.union1d(grupo_a,grupo_b))

#diferença
diferença = grupo_a - grupo_b
print(diferença)
```

```python
x_max = notas.max()
x_min = notas.min()

normalizadas = (notas - x_min) / (x_max - x_min)

print(normalizadas)
```

```python
import cv2
import random as r

cap = cv2.VideoCapture('babydance.mp4')

font = cv2.FONT_HERSHEY_SIMPLEX

while True:
    
    ret,frame = cap.read()
    
    frame = cv2.resize(frame, (500,500))
    
    if ret == False:
        break
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    cv2.putText(frame, 'Dance baby, Dance!', (r.randint(1,),r.randint(1,501)), font, 1, (0,255,150), 4, cv2.LINE_AA)
    
    cv2.imshow('babydance', frame)
    cv2.waitKey(25)
    
cap.release()
cv2.destroyAllWindows()
    
```

```python
import os

pasta = input('pasta:')

atual = os.path.abspath('')

if not os.path.isdir(pasta):
    os.mkdir(pasta)
    print('pasta criada com sucesso')
else:
    print('pasta já existente...')
    
arq = input('arquivo:')

arq = arq + '.txt'

if not os.path.isdir(arq):

    with open(arq, 'w') as f:
        f.write('')
    print('arquivo criado')
    origem = os.path.join(atual,arq)
    destino = os.path.join(atual,pasta,arq)

    os.rename(origem,destino)

    print(arq + ' movido para ' + pasta)
    
else:
    print('arquivo já existente')
```

```python
from datetime import datetime
import json

semana = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']

data = datetime.now()

dia_nome = semana[data.weekday()]

final = f'{data.day}-{data.month}-{data.year}'

final = json.dumps({final:dia_nome})

print(final)
```
