```python
tips = pd.read_csv('CSV/tips.csv',sep=';')
tips.head()
sns.relplot(x='total_bill', y='tip', hue = 'day', data = tips)
```

```python
# sns.relplot() -> Cria grfico relacional (scatter por padro)
# x='total_bill' -> Define varivel do eixo X (valor total da conta)
# y='tip' -> Define varivel do eixo Y (valor da gorjeta)
# hue='day' -> Diferencia os pontos por cor (dia da semana)
# col='time' -> Divide o grfico em colunas (almoo/jantar)
# row='sex' -> Divide o grfico em linhas (sexo)
# data=tips -> Define o dataset utilizado

sns.relplot(x='total_bill', y='tip', hue = 'day', col = 'time', row = 'sex', data = tips)
```

```python
# k = ['Thur','Fri','Sat','Sun']
# Lista com a ordem correta dos dias da semana

# dayWeek = sorted(tips.day, key=k.index)
# Ordena a coluna 'day' de acordo com a lista k
# (define a ordem personalizada para os dias)

# sns.relplot(..., kind='line')
# Cria grfico relacional do tipo linha

# x='day'
# Eixo X  dias da semana

# y='total_bill'
# Eixo Y  valor total da conta

# data=tips
# Dataset utilizado

k = ['Thur','Fri','Sat','Sun']
dayWeek = sorted(tips.day, key=k.index) # para o grfico ficar na ordem certa
# caso contrrio, a ordem  a em que os itens aparecem no dataset
sns.relplot(x='day',y='total_bill',data=tips,kind='line')
```

```python
# sns.relplot() -> Cria grfico relacional
# kind='line' -> Define grfico de linha

# x=dayWeek
# Define os dias da semana j ordenados como eixo X

# y='total_bill'
# Eixo Y  valor total da conta

# hue='sex'
# Diferencia as linhas por cor (sexo)

# style='sex'
# Diferencia tambm pelo estilo da linha (trao/marcador) por sexo

# data=tips
# Dataset utilizado

sns.relplot(x=dayWeek,y='total_bill',data=tips,kind='line',hue='sex',style='sex')
```

```python
# sns.catplot() -> Cria grfico para dados categricos
# (padro: kind='strip')

# x=dayWeek
# Define os dias da semana ordenados no eixo X

# y='total_bill'
# Eixo Y  valor total da conta

# hue=dayWeek
# Colore os pontos de acordo com o dia da semana
# (cada dia ter uma cor diferente)

# data=tips
# Dataset utilizado
sns.catplot(x=dayWeek, hue = dayWeek, y='total_bill', data=tips)
```

```python
# sns.catplot(..., kind='box')
# Agora o grfico  um BOXPLOT (no  mais strip)

# x=dayWeek
# Dias da semana no eixo X

# y='total_bill'
# Valor total da conta no eixo Y

# hue=dayWeek
# Colore cada box pelo dia (aqui  redundante)

# kind='box'
# Mostra distribuio estatstica:
# - Mediana
# - Quartis (Q1 e Q3)
# - Valores mnimos e mximos
# - Possveis outliers
sns.catplot(x=dayWeek, y='total_bill',hue = dayWeek, data=tips, kind='box')
```

```python
# sns.catplot() -> grfico para dados categricos
# kind='boxen' -> Boxenplot: verso detalhada do boxplot
#                mostra melhor a distribuio, principalmente caudas/outliers
# x=dayWeek -> categorias no eixo X (dias da semana)
# y='total_bill' -> valores numricos no eixo Y (total da conta)
# hue=dayWeek -> colore os boxes por dia (redundante aqui, normalmente usar outra varivel)
# data=tips -> dataframe utilizado

sns.catplot(x=dayWeek, y='total_bill',hue = dayWeek,  data=tips, kind='boxen')
```

```python
# sns.catplot() -> grfico para dados categricos
# kind='violin' -> Violin plot: mostra a distribuio dos dados de forma suavizada
#                  similar ao boxplot, mas com densidade da distribuio
# x=dayWeek -> categorias no eixo X (dias da semana)
# y='total_bill' -> valores numricos no eixo Y (total da conta)
# hue=dayWeek -> colore os violinos por dia (aqui  redundante)
# data=tips -> dataframe utilizado

sns.catplot(x=dayWeek, y='total_bill', data=tips, kind='violin',hue = dayWeek)
```

```python
# distplot -> histograma + curva de densidade
# Mostra a distribuio de uma varivel numrica
# Ex: total da conta no dataset tips

sns.distplot(tips['total_bill'])
```

```python
# displot -> histograma de uma varivel numrica
# kde=True -> adiciona curva de densidade
# Mostra a distribuio da varivel
# Dataset: tips['total_bill']

sns.displot(tips['sex'], kde=True)
```

```python
# jointplot -> grfico combinado de duas variveis numricas
# Centro: scatter plot
# Margens: histogramas ou KDE
# Permite ver relao + distribuio individual
# Dataset: tips

sns.jointplot(x='total_bill',y='tip',data=tips)
```
