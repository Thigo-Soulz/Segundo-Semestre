```python
import seaborn as sns
import pandas as pd

# catplot -> gráfico para variáveis categóricas
# kind='bar' -> gráfico de barras mostrando valores numéricos
# x='passengers' -> valores a serem exibidos
# y='month' -> categorias
# data=flights -> dataset utilizado

flights = pd.read_csv('CSV/flights.csv',sep=',')
filtro = flights['year'] == 1950
flights = flights[filtro]

sns.catplot( x = "passengers", y= "month",kind= "bar", data=flights, hue = 'month')
```
