```python
import seaborn as sns
import pandas as pd

# catplot -> gráfico para variáveis categóricas
# kind='bar' -> gráfico de barras mostrando valores numéricos
# x='passengers' -> valores a serem exibidos
# y='month' -> categorias
# data=tips -> dataset utilizado
tips = pd.read_csv('CSV/flights.csv',sep=',')
sns.catplot(x = "passengers"y= "month",data=tips, kind='count')
```
