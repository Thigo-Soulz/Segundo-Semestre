```python
tips = pd.read_csv('CSV/tips.csv',sep=';')
tips.head()
sns.relplot(x='total_bill', y='tip', hue = 'day', data = tips)
```

```python
sns.relplot(x='total_bill', y='tip', style= 'smoker', data = tips)
```

```python
sns.relplot(x='total_bill', y='tip', style = 'smoker', hue = 'day', size = 'servings', data = tips)
```

```python
sns.relplot(x="total_bill", y="tip", col="time",
hue="time", size="servings", style="sex",
palette=["b","g"], sizes=(10, 100),data=tips)
```

```python
# sns.relplot() -> Cria grfico relacional (scatter por padro)
# x='total_bill' -> Define varivel do eixo X (valor total da conta)
# y='tip' -> Define varivel do eixo Y (valor da gorjeta)
# hue='day' -> Diferencia os pontos por cor (dia da semana)
# col='time' -> Divide o grfico em colunas (almoo/jantar)
# row='sex' -> Divide o grfico em linhas (sexo)
# data=tips -> Define o dataset utilizado

sns.relplot(x='total_bill', y='tip', hue = 'day', col = 'time', row = 'sex', data = tips)
```
